```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Uploading csv
df = pd.read_csv(r"/Users/soum/Documents/UNCC/DTSC/CSV class work/unemployed_population_1978-12_to_2023-07.csv")
df2 = pd.read_csv(r"/Users/soum/Documents/UNCC/DTSC/CSV class work/average unemployment.csv")

# Convert the 'date' column to datetime                                       
df['date'] = pd.to_datetime(df['date'])   
df2['date'] = pd.to_datetime(df2['date'])                                   
                                               
# Filter the DataFrame for the years 1988 to 1998                             
df_1988_1998 = df[(df['date'].dt.year >= 1988) & (df['date'].dt.year <=        
1998)]                                                                 
df2_1988_1998 = df2[(df2['date'].dt.year >= 1988) & (df2['date'].dt.year <=        
1998)]           
```

```python
# Setting up our columns
dates = df_1988_1998.iloc[:,0]
dates2 = df2_1988_1998.iloc[:,0]

w = df_1988_1998.iloc[:,11]
w_16_24 = df_1988_1998.iloc[:,12]
w_25_54 = df_1988_1998.iloc[:,13]

m = df_1988_1998.iloc[:,21]
m_16_24 = df_1988_1998.iloc[:,22]  
m_25_54 = df_1988_1998.iloc[:,23]  

w_hs = df_1988_1998.iloc[:,17]
w_sc = df_1988_1998.iloc[:,18]
w_bd = df_1988_1998.iloc[:,19]
w_ad = df_1988_1998.iloc[:,20]

m_hs = df_1988_1998.iloc[:,27]
m_sc = df_1988_1998.iloc[:,28]
m_bd = df_1988_1998.iloc[:,29]
m_ad = df_1988_1998.iloc[:,30]

avg_w_age = df2_1988_1998.iloc[:,23]
avg_m_age = df2_1988_1998.iloc[:,24]

summary_w_columns = df2_1988_1998[['women_age_average', 'women_high_school', 'women_bachelors_degree', 'women_advanced_degree']]
summary_m_columns = df2_1988_1998[['men_age_average', 'men_high_school', 'men_bachelors_degree', 'men_advanced_degree']]
```

```python
# Creating a line chart for Women/Men Age
fig, ax = plt.subplots()
ax.plot(dates,w_16_24)
ax.plot(dates,m_16_24)
ax.plot(dates,w_25_54)
ax.plot(dates,m_25_54)

ax.set_title("Line graph of Womens/Mens Unemployment (Age)")
ax.legend(["Women 16-24", "Men 16-24", "Women 25-54", "Men 25-54"])
ax.set_xlabel("year")
ax.set_ylabel("Unemployment Rate (%)")


plt.show()

```

```python
# Creating a line chart for Average Women/Men Age
fig, ax = plt.subplots()
ax.plot(dates2,avg_w_age)
ax.plot(dates2,avg_m_age)

ax.set_title("Line graph of Average Womens/Mens Unemployment (Age)")
ax.legend(["Women 16-54", "Men 16-54"])
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Rate (%)")


plt.show()
```

```python
# Creating a line chart for Womens education with men hs
fig, ax = plt.subplots()
ax.plot(dates,w_hs)
ax.plot(dates,w_sc)
ax.plot(dates,w_bd)
ax.plot(dates,w_ad)
# ax.plot(dates,m_hs)


ax.set_title("Line graph of Womens Unemployment (Education)")
ax.legend(["Highschool", "Some College", "Bachelors Degree", "Advanced Degree"])
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Rate (%)")


plt.show()
```

```python
# Creating a line chart for Mens education
fig, ax = plt.subplots()

ax.plot(dates,m_hs)
ax.plot(dates,m_sc)
ax.plot(dates,m_bd)
ax.plot(dates,m_ad)

ax.set_title("Line graph of Mens Unemployment (Education)")
ax.legend(["Highschool", "Some College", "Bachelors Degree", "Advanced Degree"])
ax.set_xlabel("Year")
ax.set_ylabel("Unemployment Rate (%)")


plt.show()
```

```python
# creating a box plot for the statistics

plt.boxplot([w_hs, m_hs], labels = ["Women Highschool", "Men Highschool"])
plt.xlabel("Gender (Highschool)")
plt.ylabel("Unemployment rate in %")

plt.show
```

```python
# creating a box plot for the statistics

plt.boxplot([avg_w_age, avg_m_age], labels = ["Avg. Women Age", "Avg. Men Age"])
plt.xlabel("Gender (Average Age)")
plt.ylabel("Unemployment rate in %")

plt.show
```
